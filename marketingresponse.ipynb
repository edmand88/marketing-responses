```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['campaign_responses.csv']))
print(df)
```

```python
df = df.drop('customer_id', axis=1)
print(df)
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
print(df)
```

```python
df['employed'] = df['employed'].map({'Yes': 1, 'No': 0})
print(df)
```

```python
df['marital_status'] = df['marital_status'].map({'Married': 1, 'Single': 0})
print(df)
```

```python
df['responded'] = df['responded'].map({'Yes': 1, 'No': 0})
print(df)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('responded', axis=1)
y = df['responded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=13, stratify=y, random_state=42)

print("X_train size:", len(X_train))
print("y_train size:", len(y_train))
print("X_test size:", len(X_test))
print("y_test size:", len(y_test))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler

# 1. Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 2. Build the neural network model
model = Sequential()
model.add(Dense(12, input_dim=X_train_scaled.shape[1], activation='relu'))  # Input layer with 12 nodes
model.add(Dense(8, activation='relu'))  # Hidden layer with 8 nodes
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 node (binary classification)

# 3. Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. Train the model
model.fit(X_train_scaled, y_train, epochs=50, batch_size=10, validation_split=0.2)

# 5. Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
```
